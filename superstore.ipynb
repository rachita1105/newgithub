```python
import matplotlib.pyplot as plt
df['sales'] = pd.to_numeric(df['sales'], errors='coerce')

# Plot sales by region using the correct column names
df.groupby('region')['sales'].sum().plot(kind='bar', title='Sales by Region', color='lightgreen', figsize=(8,5))
plt.xlabel("region")
plt.ylabel("Total Sales")
plt.tight_layout()
plt.show()
```

```python
df['order_date'] = pd.to_datetime(df['order_date'], dayfirst=True, errors='coerce')
```

```python
df['Month'] = df['order_date'].dt.to_period('M').dt.to_timestamp()
```

```python
monthly_sales = df.groupby('Month')['sales'].sum()
```

```python
import matplotlib.pyplot as plt

monthly_sales.plot(kind='line', figsize=(10, 6), title='Monthly Sales Trend')
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df.groupby('category')['profit'].sum().plot(kind='bar', title='Profit by Category')
```

```python
monthly_sales.plot(kind='pie', figsize=(50, 6), title='Monthly Sales Trend')
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.grid(True)
plt.tight_layout()
plt.xticks(rotation=50)
plt.show()
```
